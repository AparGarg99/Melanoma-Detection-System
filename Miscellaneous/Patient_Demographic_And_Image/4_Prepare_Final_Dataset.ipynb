```python
import pandas as pd
import os
import glob
```

```python
df11 = pd.read_csv('processed_data_ham.csv')
df11.shape
```

```python
df11.head()
```

```python
# one-hot encoding
df12 = df11.copy()
for i in df12.columns:
    if(df12[i].dtype!='float64'):
        df12[i]=pd.get_dummies(df12[i]).iloc[:,0:1]
df12.head()
```

```python
df12['Name']=df11['Name']
df1=df12.copy()
df1.head()
```

```python
df1['Patient_Age']=df1['Patient_Age']/df1['Patient_Age'].mean()
```

```python
df1.head()
```

```python
path = r'Extracted Features'                     
all_files = glob.glob(os.path.join(path + "/*.csv"))

df_from_each_file = (pd.read_csv(f) for f in all_files)
df2 = pd.concat(df_from_each_file, ignore_index=True)
```

```python
df2
```

```python
df2 = df2.dropna()
df2 = df2.drop_duplicates(subset=['Name'])
df2.shape
```

```python
df = pd.merge(df1,df2,on='Name')
df.head()
```
