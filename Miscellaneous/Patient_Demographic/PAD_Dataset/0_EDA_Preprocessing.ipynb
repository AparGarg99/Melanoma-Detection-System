## AutoViz

Documentation: https://pypi.org/project/autoviz/

```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
sep = ","
dft = AV.AutoViz(
    filename='',
    sep=",",
    depVar="",
    dfte=df,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```

Univariate and Bi-variate Analysis for columns with <40% missing values:
* 56.2% (86.5% excluding missing values) of the patients don't smoke.
* 49% (75.4% excluding missing values) of the patients don't drink.
* Parents of patients are mainly from POMERANIA, GERMANY, ITALY.
* Patients are mostly old (more than 50 years).
* Ratio of male to female patients is same.
* More than 50% of lesions occur in face, forearm and chest.
* Most of the lesions have diameter in the range 1-20 mm.
* Only 2.3% of the patients suffer from melanoma.
* Around 63% of the patients reported itching in skin lesion.
* Lesion did not hurt for more than 80% of the patients.
* Lesion did not recently change in more than 70% of the patients.
* Approximately 26% patients reported bleeding from lesion.
* diameter1 and diameter2 are highly correlated -> lesion is usually circular in shape.
