```python
import pandas as pd
```

```python
df = pd.read_csv('PAD_dataset/metadata.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
l=['smoke', 'drink', 'background_father',
       'background_mother', 'age', 'pesticide', 'gender',
       'skin_cancer_history', 'cancer_history','has_piped_water','has_sewage_system','diagnostic']
```

```python
df=df[l]
```

```python
df
```

```python
df = df.dropna(how='any')
```

```python
df['Melanoma']=df['diagnostic'].apply(lambda x:'False' if x!='MEL' else 'True')
```

```python
df=df.drop('diagnostic',axis=1)
```

```python
df = df.drop_duplicates()
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_csv('processed_data_pad.csv',index=False)
```
